```python
formula = 'WS2'
xc = 'PBE'
out_dir = './out/'
name = out_dir + 'gs_'+ formula + '_xc=' + xc + '_bs.gpw'

from gpaw import GPAW, PW, FermiDirac
from ase.io import read
```

```python
structure = read('../structures/' + formula + '.json')
kpath = structure.cell.bandpath(path='GM', npoints=10, pbc=structure.pbc, eps=0.01)
```

```python
# Restart from ground state and fix potential:
calc = GPAW(name)
```

```python
from ase.dft.kpoints import monkhorst_pack

calc.fixed_density(
    nbands=16,
    symmetry='off',
    kpts=kpath,
    convergence={'bands': 6})
```

```python
bs = calc.band_structure()
bs.plot(show=True)
# bs.plot(filename='bandstructure.png', show=True, emax=10.0)
# bs.plot(filename='bandstructure.pdf', show=True, emax=10.0)
```
